```python
#import packages
#yfinance is an api used to use stock prices and so...
import yfinance as yf
import pandas as pd
```

```python
# Ticker is a module of yfinance that is used to acces stocks of particular stocks...
apple = yf.Ticker('AAPL') #AAPL is stock symbol of apple stock..
# Here we basically created an apple stock as an object.
```

```python
# In Ticker each stock info is stored in form of JSON, so we need to import json module to work with them.
import json
```

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json
#Here Wget is used to download the apple file from web into our current working directory.
#wget stands for "World Wide Web Get". It is a command-line tool that retrieves files from the web using HTTP, HTTPS, or FTP protocols.
```

```python
with open('apple.json') as json_file:
  apple_info = json.load(json_file) #json.load converts json into mostly python dict..
type(apple_info)
apple_info
```

```python
#This is apple_info is basically like a python dictionary. We can get country name using country.
apple_info['country']
```

```python
#Using the history() method we can access the price of apple stock for any period of time .
apple_share_price_data = apple.history(period="max")
```

```python
apple_share_price_data
```

```python
apple_share_price_data.reset_index(inplace = True)
#Normally the index values are dates, but by reseting index we created index values as numbers...
```

```python
apple_share_price_data
```

```python
apple_share_price_data.plot(x="Date", y="Open")
```
